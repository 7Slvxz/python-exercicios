FAZER UM PROGRAMA PARA LER AS MEDIDAS DA BASE E ALTURA DE UM RETANGULO. EM SEGUIDA, MOSTRAR O VALOR DA ÁREA, PERIMETRO E DIAGONAL DESTE RETANGULO, COM QUATRO CASAS DECIMAIS, CONFORME EXEMPLOS.

Exemplo - 1

*  Base do retangulo : 4.0
*  Altura do retangulo : 5.0
* Area = 20.0000
*  Perimetro = 18.0000
* Diagonal = 6.4031


Exemplo - 2

*   Base do retangulo : 10.3

*   Alturado retangulo : 13.1
*   Area = 134.9300


*   Perimetro = 46.80000

*   Diagonal = 16.6643





```python
import math

base = float(input("Digite a base do retangulo: "))
altura = float(input("Digite a altura do retangulo: "))


resultadoArea = base * altura
resultadoPerimetro = 2 * (base + altura)
resultadoDiagonal = math.sqrt(base**2 + altura**2)

print ("A area do triangulo é igual a: ",resultadoArea)
print ("O perimetro do triangulo é igual a: ",resultadoPerimetro)
print ("A diagonal do triangulo é igual a: ",resultadoDiagonal)
```

FAZER UM PROGRAMA PARA LER O NOME E IDADE DE DUAS PESSOAS. AO FINAL MOSTRAR UMA MENSAGEM COM OS NOMES E A IDADE MÉDIA ENTRE ESSAS PESSOAS, COM UMA CASA DECIMAL, CONFORME EXEMPLO.

Exemplo

*  Dados da primeira pessoa:
*   Nome: Maria Silva
*   Idade: 19

*   Dados da segunda pessoa:
*   Nome: Joao Melo
*   Idade: 20


*   A idade media de Maria Silva e Joao Melo é de ??









```python
nome1 = input("Digite o nome da primeira pessoa: ")
idade1 = int(input("Digite a idade da primeira pessoa: "))

nome2 = input("Digite o nome da segunda pessoa: ")
idade2 = int(input("Digite a idade da segunda pessoa: "))

media = (idade1 + idade2) / 2
# PARA CALCULAR A MEDIA DE DOIS NUMEROS USE a = (n1 + n2) / 2
print("A idade media de", nome1,"e", nome2, "é de", media, "anos")
```

Fazer um programa para calcular o troco no processo de pagamento de um produto de uma mercearia.
O programa deve ler o preço unitário do produto, a quantidade de unidades compradas deste produto,
e o valor em dinheiro dado pelo cliente (suponha que haja dinheiro suficiente). Seu programa deve
mostrar o valor do troco a ser devolvido ao cliente.

* Exemplo 1:

Preço unitário do produto: 8.00

Quantidade comprada: 2

Dinheiro recebido: 20.00

TROCO = 4.00

* Exemplo 2:

Preço unitário do produto: 30.00

Quantidade comprada: 3

Dinheiro recebido: 100.00

TROCO = 10.00


```python
preco1 = float(input("Digite o preço unitário do produto: "))
quant1 = int(input("Digite a quantidade comprada: "))
receber = float(input("Digite o dinheiro recebido:"))

troco = receber - (preco1 * quant1)
print("O troco deve ser de: ", troco, "reais")
```
