```python
total = 0
# Initialize num to something that allows the loop to start, like 0.
# The first input will then set num.
num = 0

while num > -1: # Loop continues as long as the last entered number was greater than -1
    num = int(input("Digite um numero (insira -1 ou um numero negativo para parar): "))
    if num > -1: # Only add to total if the number is positive or zero
        total += num
        print("A soma parcial é ", total)

print("A soma total final dos números inseridos (excluindo o valor negativo que parou a contagem) é ", total)
```

```python
from random import randint
eleitor = 0
cand1 = 0
cand2 = 0
cand3= 0
while eleitor > 100000:
  voto = randint(1,3)
  if voto == 1:
    cand1 += 1
  elif voto == 2:
    cand2 += 1
  else voto == 3:
    cand3 += 1
    eleitor += 1
print("Candidato 1: ", cand1)
print("Candidato 2: ", cand2)
print("Candidato 3: ", cand3)
```

```python
from random import randint
chute = 0
escolha = randint(1,100)
while chute!= escolha:
  chute = int(input("Digite um valor entre 1 e 100: "))
  if chute > escolha:
    print("Errou! Tente algo mais baixo")
  if chute < escolha:
    print("Errou! Tente algo mais alto")
  if chute == escolha:
    print("Parabens! Você acertou!")
```
